# Creation of YOLO Dataset from Image Processing Pipeline

### Extracting ellipse parameters, cleaning data, and generating height vs. width

Following tasks are performed:
- Train YOLO Model on annotated data
- Compare detection rates
- save model


```python
# Import necessary libraries for YOLO training and data processing
import numpy as np                     # Numerical operations and array handling
from ultralytics import YOLO           # YOLO model for object detection
import matplotlib.pyplot as plt        # Plotting and visualization
from PIL import Image                  # Image handling and manipulation
import pandas as pd                    # Data analysis and manipulation
import torch                            # PyTorch, used by YOLO for deep learning


```

```python

# Load a pre-trained YOLO model (you can choose n, s, m, l, or x versions)
model = YOLO("yolo11n.pt")

# Start training on your custom dataset
model.train(data="cone.yaml", epochs=100, imgsz=256, workers = 4)
```

```python

# Load the best model if `model` is not defined
if 'model' not in globals():
    # Adjust the path to your saved best model
    model = YOLO("/pn_local/2_1/runs/detect/train62/weights/best.pt")

zero_or_one = []
im_list = np.load('im_list.npy')  # Load the saved list of images
total_detections = 0

for idx, im_array in enumerate(im_list):
    # Ensure the input image is in grayscale
    if len(im_array.shape) == 2:  # If only 2 dimensions, it's grayscale
        # Expand the grayscale image to 3 channels (RGB)
        rgb_array = np.stack((im_array,) * 3, axis=-1)
    elif len(im_array.shape) == 3 and im_array.shape[2] == 1:
        # If the array has a single channel in the 3rd dimension, replicate it
        rgb_array = np.concatenate([im_array] * 3, axis=-1)
    else:
        # If the image is already RGB, leave it as is
        rgb_array = im_array

    # Convert NumPy array to a PIL Image object
    rgb_image = Image.fromarray(rgb_array)

    # Pass the RGB image to the YOLO model for detection
    results = model(rgb_image, verbose=False)

    for i, r in enumerate(results):
        # Count the number of detections in this image
        num_detections = len(r.boxes)
        if num_detections in [0, 1]:
            zero_or_one.append(idx)  # Track images with 0 or 1 detection
        total_detections += num_detections

        # Visualize detections (convert BGR to RGB)
        im_bgr = r.plot()
        im_rgb = Image.fromarray(im_bgr[..., ::-1])
        # Optional plotting (commented out)
        # plt.figure(figsize=(4, 4))
        # plt.imshow(im_rgb)
        # plt.axis('off')
        # plt.show()

        # Optional saving results to disk
        # r.save(filename=f"results{i}.jpg")

# Print detection rate per image and per object (assuming 3 cones per image)
print("Detection rate: " + str(total_detections / (len(im_list) * 3)))
```

```python
# Load the cleaned ellipse data
df_ellips_cleaned = pd.read_csv('output.csv', header=0)

# Select the first three columns (cone widths)
first_three_columns = df_ellips_cleaned.iloc[:, :3]

# Count the non-NaN entries in these columns
non_nan_counts = first_three_columns.notna().sum()

# Sum across the three columns to get the total number of detected cones
total_non_nan_count = non_nan_counts.sum()

print("Total number of non-NaN entries in the first three columns:", total_non_nan_count)
# Compute detection rate assuming 3 cones per image
print("Detection rate: " + str(total_non_nan_count / (len(im_list) * 3)))
```

```python
# If the detection rate is higher than 98%, export the model
if total_detections / (len(im_list) * 3) > 0.98:
    # Export the model in YOLO format (for deployment or inference)
    model.export()

    # Save the raw PyTorch model to a .pth file
    torch.save(model.model, "/pn_local/2_1/best_model.pth")
```
